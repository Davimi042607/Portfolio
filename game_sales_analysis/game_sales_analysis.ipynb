Аббревиатура tbd от английского To Be Determined (будет определено) или To Be Decided (будет решено). Используется, если какая-то информация еще не определена или решение по вопросу не принято. #### Столбец 'year_of_release'.

Не смотря на незначительное количество пропусков в количестве 269, попробуем заполнить данные значения на основании названии игры и платформы. В основном выпуск игры осуществляется на все платформы одновременно, соответственно имеющийся год игры на другой платформе используем для замены пропуска. Выведим строки с пропущенными данными в столбце.

````python
data.query('year_of_release.isna()')
````

Из среза выше выбирем игру 'Madden NFL 2004' и выведим строки с этой игрой в срезе по всему датасету.

````python
data[data['name'] == 'Madden NFL 2004']
````

Игра 'Madden NFL 2004' вышла на платформах XB, GC, GBA в 2003 году, логично что и на PS2 это произошло в 2003году. Заменим пропуски в этом столбце на основании групировки по имени игры, выбирать будем максимальное значение, т.к. возможно что платформа появилась позже самой игры.

````python
data['year_of_release'] = (
    data['year_of_release'].
    fillna(data.groupby('name')['year_of_release'].transform('max'))
)
````

Проверим результат на примере все той же игры 'Madden NFL 2004'

````python
data[data['name'] == 'Madden NFL 2004']
````

Также игры вышешедшие в период 1980-1990, имеющие маленькое количество продаж, могли просто не оцениваться пользователями  в силу отсутствия обратной связи со стороны пользователей, а со стороны критиков остаться не замеченными. #### Столбец со значениями 'rating'.

Выведим уникальные значения.

````python
data['rating'].unique()
````

- «E» («Everyone») — «Для всех»
- nan
- «M» («Mature») — «Для взрослых»
- «T» («Teen») — «Подросткам»
- «E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»
- K-A - for Kids to Adults
- «AO» («Adults Only 18+») — «Только для взрослых»
- «EC» («Early childhood») — «Для детей младшего возраста»

Рейтинг К-А устарел и на данный момент не используется. По характеристикам соответствует рейтингу E. Произведем замену К-А на Е.
````python
data.loc[data['rating'] == 'K-A', 'rating'] = 'E'
data['rating'] = data['rating'].fillna('Unknown')
````

````python
data.pivot_table(
    index=['year_of_release'],
    columns='platform',
    values='total_sales',
    aggfunc='sum').plot(
                        linewidth=3,
                        alpha=0.7,
                        title='Количество копии проданных игр на различные платформы в год',
                        xlabel='Год',
                        ylabel='Количество копии, млн.',
                        grid=True,
                        legend=True,
                        figsize=(16, 10)
);
````

Средняя продолжительность работы платформы 5-10 лет. Для дальнейшего анализа устаревшие платформы не нужны, сделаем срез для выпуска игр и продажи копии с 2013 года и построим график.

````python
data.query('year_of_release > 2013').pivot_table(
    index=['year_of_release'],
    columns='platform',
    values='name',
    aggfunc='count').plot(
                        linewidth=3,
                        alpha=0.7,
                        title='Количество выпущенных игр на различные платформы в год',
                        xlabel='Год',
                        ylabel='Количество игр',
                        grid=True,
                        legend=True,
                        figsize=(12, 8)
);
````

````python
data.query('year_of_release >= 2013').pivot_table(
    index=['year_of_release'],
    columns='platform',
    values='total_sales',
    aggfunc='sum').plot(
                        linewidth=3,
                        alpha=0.7,
                        title='Количество проданных копии игр на различные платформы в год',
                        xlabel='Год',
                        ylabel='Количество копии игр, млн.',
                        grid=True,
                        legend=True,
                        figsize=(12, 8)
);
````

**Согласно графиков выпуска и продаж игр наблюдаем следующие перспективные платформы для анализа: PSV год выпуска 2011, WiiU год выпуска 2012, XOne год выпуска 2013, PS4 год выпуска 2013. Также рассмотрим динамику выпуска и продаж игр в период с 2013 по 2015 год включительно, построив графики (2016 не рассматриваем из-за отсутствия полноты данных).**

## Проверка гипотез

Применим критерий Стьюдента для проверки равентства средних значений в двух выборках.

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сохраним под переменной xone_rating оценки пользователей платформы XOne. Выведим размер всей совокупности.

````python
xone_rating = data.query('platform == "XOne"')['user_score'].dropna()
xone_rating.count()
````

Под перевменной pc_rating сохраним оценки пользователей платформы PC. Выведим также размер всей совокупности.

````python
pc_rating = data.query('platform == "PC"')['user_score'].dropna()
pc_rating.count()
````

Т.к. размер выборок разный и неизвестно равенство дисперсий применим обязательный параметр equal_var = False.

Сформулируем нулевую гипотезу H_0 и альтернативную H_a. Уровень статистической значимости обозначим в 5%.
```
H_0: Пользовательский рейтинг (user_score) платформы "Xbox One" = пользовательскому рейтингу (user_score) платформы "PC"`
H_a: Пользовательский рейтинг (user_score) платформы "Xbox One" ≠ пользовательскому рейтингу (user_score) платформы "PC"`
alpha = 0.05
```

````python
alpha = 0.05

results = st.ttest_ind(xone_rating, pc_rating, equal_var=False)
print(results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

**Средние пользовательские рейтинги платформ Xbox One и PC разные**

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные

Совокупности независимые, применяем метод scipy.stats.ttest_ind().

Сохраним под переменной action_rating оценки пользователей жанра Action. Выведим на экран размер.

````python
action_rating = data.query('genre == "Action"')['user_score'].dropna()
action_rating.count()
````

Сохраним под переменной sports_rating оценки пользователей жанра Sports. Выведим на экран размер.

````python
sports_rating = data.query('genre == "Sports"')['user_score'].dropna()
sports_rating.count()
````

Т.к. размер выборок разный и неизвестно равенство дисперсий применим обязательный параметр equal_var = False.

Сформулируем нулевую гипотезу H_0 и альтернативную H_a. Уровень статистической значимости обозначим в 5%.

```
H_0: Пользовательский рейтинг (user_score) платформы "Xbox One" = пользовательскому рейтингу (user_score) платформы "PC"`
H_a: Пользовательский рейтинг (user_score) платформы "Xbox One" ≠ пользовательскому рейтингу (user_score) платформы "PC"`
alpha = 0.05
```

````python
alpha = 0.05

results = st.ttest_ind(action_rating, sports_rating, equal_var=False)
print(results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

**Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые**

Пятерку лидеров в обоих регионах закрывает платформа 3DS, которая в свою очередь является самой популярной в Японии. PS3, PSV, PS4 занимают соответственно 2, 3, 4 место с ущественным отстованием от 3DS и незначительной разницей между сабой. На пятом месте в Японии платформа WiiU.**        

**Предпочтения по жанрам также схоже в Европе и Америке, с единственной разницей в жанре на пятом месте по популярности, Misc - Америка, Racing - Европа. Action и Shooter занимают первое и второе место и составляют приблизительно суммарно 66% продаж в пятерке самых популярных. Shooter и Role-Plaing занимают третье и четвертое место. В японии лидер по популярности - Role-Playing, ~45%. На втором месте Action с долей ~35%. Далее Misc, Fighting и Shooter приблизительно с одинаковыми показателями продаж.** 

**Для америки и Европы в плане рейтинга игр также наблюдается схожесть предпочтений. На первом месте игры для взрослых, второе и  третье, игры с неизвестным рейтингом и для всех. Четвертое и пятое - игры для подростков и с ограничением по возрасту до 10 лет, с одинаковыми показателями продаж. В Японии на первом месте игры с неизвестным рейтингом, на втором игры для подростков, игры для взрослых и для всех занимают соответсвенно 3 и 4 место. На пятом с ограничением до 10лет.**

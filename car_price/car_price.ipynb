### Предобработка данных

Для предсказания стоимости автомобиля необходимо учитывать множество факторов, начиная от технических характеристик до состояния, но в первую очередь - это чему предсказываем, т.е. бренд - модель и даже наверное модель здесь имеет более ключевое значение. Однако в признаке модель помимо пропусков в количестве 6% присутствует тип кузова 'other', который является заглушкой не представляющей возможности определить конкретный автомобиль. Данные объекты необходимо будет удалить. Столбцы бренд и модель объединить в один. 
Пропуски в столбцах тип кузова, коробка, топливо заполним по признакам бренд-модель-год, пропуски в столбце о наличие ремонта - заглушкой unknown. Анамольные значения:
- Price является целевым признаком. Значения приближеные и равные 0 - удалить,
- RegistrationYear - год выпуска является также ключевым для определения стоимости конкретной модели, поэтому восстановление аномальных значений по среднему или медиане не применимо, требуется удаление. После очистки объединить со столбцом RegistrationMonth и вывести новый признак - возраст автомобиля. 
- Power - аномальные значения заполним средним по моделе.

Столбец NumberOfPictures заполнен константым значением, удаляем. Столбец PostalCode - является индивидуальным признаком, не представляющем информативности для моделей, удаляем. Столбцы с датами также не представляют ценности, удаляем.

Приведем к соответствующему формату названия столбцов датафрейма

```python
df = clean_columns(df, case = 'snake')
df.columns.tolist()
```

#### Модель - бренд

Выведем бренды соответствующие каждой моделе.

```python
for name in df['model'].unique():
    print(name, df.loc[df['model'] == name, 'brand'].unique())
```

В основном модель соответствует одному бренду. В незначительном количестве присутствуют одинаковые модели у разных брендов (пример: grand ['jeep' 'suzuki' 'chrysler'], matiz ['chevrolet' 'daewoo'] и т.п.). При создании нового признака бренд-модель решим эту проблему. Модель other присутствующую во многих брендах удалим.

Также выявлены следующие моменты, требующие обработки:
- модель Range Rover записана как rangerover и range_rover
- модели discovery, freelander, defender указан бренд rover, чего не может быть, т.к. эти модели есть только у land_rover

Выведем модели соответствующие каждому бренду

```python
for name in df['brand'].unique():
    print(name, df.loc[df['brand'] == name, 'model'].unique())
    print()
```

Два момента:
- sonstige_autos [nan] - данный бренд не имеет моделей, необходимо сохранить.
- rover ['other' 'discovery' nan 'rangerover' 'freelander' 'defender'] - четыре модели 'rover discovery', 'rangerover', 'freelander', 'defender' относятся к бренду land_rover. 

```python
df.loc[df['model'] == 'rangerover', 'model'] = 'range_rover'
```

```python
for name in ['discovery', 'range_rover', 'freelander', 'defender']:
    df.loc[df['model'] == name, 'brand'] = 'land_rover'
```

```python
df.loc[df['brand'] == 'rover', 'model'].unique()
```

Для бренда rover переименуем other и nan в unknown. Для бренда sonstige_autos также в unknown. Остальные модели other удалим.

```python
df.loc[df['brand'] == 'rover', 'model'] = 'unknown'
df.loc[df['brand'] == 'sonstige_autos', 'model'] = 'unknown'

df = df[df.model != 'other']
df.shape
```

Удалим строки с пропусками в столбце 'model'

```python
df.dropna(subset=['model'], inplace=True)
df.shape
```

```python
df['brand_model'] = df['brand'] + '_' + df['model']
df.drop(['model', 'brand'], axis=1, inplace=True)
df.head()
```

### LightGBM

Моделью LightGBM предусмотрено обучение на датасетах с категориальными признаками. Для этого необходимо изменить тип столбцов на 'category'.

```python
cat_features = ['vehicle_type', 'gearbox', 'fuel_type', 'repaired', 'brand_model']
X_train_category = X_train.copy()
X_valid_category = X_valid.copy()
X_test_category = X_test.copy()

for col in cat_features:
    for selection in [X_train_category, X_valid_category, X_test_category]:
        selection[col] = selection[col].astype('category')
   
X_valid_category.info()
```

```python
%%time

train_dataset = lgb.Dataset(
    X_train_category, y_train, categorical_feature = cat_features
)

valid_dataset = lgb.Dataset(
    X_valid_category, y_valid, categorical_feature = cat_features, reference=train_dataset
)

booster = lgb.train( 
    {"objective": "regression"}, 
    train_set=train_dataset,
    valid_sets=(valid_dataset,),
    num_boost_round=1000
)
```

```python
%%time

prediction_light = booster.predict(X_valid_category)
round((mean_squared_error(prediction_light, y_valid))**.5, 2)
```

Обучение на оригинальной выборке дает результат 1474.19

```python
results['LightGBM'] = [
    round(((mean_squared_error(prediction_light, y_valid))**0.5), 2), 
    '2.84s',
    '548ms'
]
```
